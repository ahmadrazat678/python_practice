```python
class Area:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def area(self):
        return self.a*self.b
```

```python
class Triangle(Area):

    def __init__(self,a,b):
        super().__init__(a,b)
    def area(self):
        x=super().area()
        x=x/2
        return x
        
        
    
```

```python
t=Triangle(78,45)
```

```python
print(t.area())
```

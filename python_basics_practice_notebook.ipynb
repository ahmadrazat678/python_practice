## Day 5
**Conditional Execution (if, elif, else, break, continue, pass)**

Conditionals decision making ke liye use hote hain. Break loop stop karta hai, continue next iteration par jata hai aur pass placeholder hota hai.

```python
x = 10
if x > 5:
    print("Greater")
if x < 5:
    print("Small")
else:
    print("Not small")
if x == 10:
    print("Ten")
elif x > 10:
    print("Big")
else:
    print("Other")
for i in range(5):
    if i == 3:
        break
for i in range(5):
    if i == 2:
        continue
    print(i)
for i in range(3):
    pass
```

## Nested Conditionals

Nested condition mein ek if ke andar doosra if hota hai. Yeh complex decisions ke liye use hota hai.

```python
age = 20
citizen = True
if age >= 18:
    if citizen:
        print("Eligible")
if age >= 18:
    if not citizen:
        print("Need citizenship")
num = 15
if num > 0:
    if num % 2 == 0:
        print("Positive even")
    else:
        print("Positive odd")
marks = 85
if marks >= 50:
    if marks >= 80:
        print("A grade")
    else:
        print("Pass")
```

## Conditional (Ternary) Expression

Ternary operator single line mein if‑else likhne ka short way hai.

```python
x = 10
result = "Positive" if x > 0 else "Negative"
age = 16
status = "Adult" if age >= 18 else "Minor"
num = 7
parity = "Even" if num % 2 == 0 else "Odd"
a, b = 5, 9
max_val = a if a > b else b
marks = 75
grade = "Pass" if marks >= 50 else "Fail"
temp = 30
weather = "Hot" if temp > 25 else "Cool"
n = 0
type_num = "Zero" if n == 0 else "Non-zero"
score = 90
msg = "Excellent" if score > 85 else "Good"
light = True
state = "On" if light else "Off"
is_logged = False
text = "Welcome" if is_logged else "Login first"
```

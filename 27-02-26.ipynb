```python
class  Employee:
    def __init__(self,name,salary):
        self.name = input(name)
        self.salary = int(input(salary))

    def bonus(self):
        if self.salary >= 5000:
            print("Employee is Eligible for Bonus.")
            return self.salary * 3
        else:
            print("Employee is not Eligible for Bonus.")
    
    def get__name(self):
        return self.name
    
    def set__name(self,name):
        self.name = name    
    
    def get__salary(self):
        return self.salary
    
    def set__salary(self,salary):
        self.salary = salary

    def display(self):
        print("Name of the Employee :",self.name)    
        print("Salary of the Employee :",self.salary)  
        print("Bonus of the Employee :",self.bonus()) 
         
class manager(Employee):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department = input(department)

    def get__department(self):
        return self.department
    
    def set__department(self,department):
        self.department = department

    def display(self):
        super().display()
        print("Department of the Manager :",self.department)
        print
manager1 = manager("Enter name of the Manager :","Enter Salary of the Manager :","Enter Department of the Manager :")
manager1.display()
```

```python
class Vehicle:
    def __init__(self,name,make_year,color):
        self.name = name
        self.make_year = make_year 
        self.color = color

    def get_name(self):
        return self.name
    
    def set_name(self,name):
        self.name = name

    def get_make_year(self):
        return self.make_year

    def set_make_year(self,make_year):
        self.make_year = make_year

    def get_color(self):
        return self.color
    
    def set_color(self,color):
        self.color = color

    def display(self):
        print("Name of the Vehicle :",self.name)
        print("Make Year of the Vehicle :",self.make_year)
        print("Color of the Vehicle : ",self.color)

#v1 = Vehicle("Toyota",2021,"Black")
#v1.display()

# Using inheritance

class Car(Vehicle):
    def __init__(self, brand, make_year, color, model, price, car_type):
        super().__init__(brand, make_year, color)
        self.model = model
        self.price = price
        self.car_type = car_type

    def get_model(self):
        return self.model

    def set_model(self, model):
        self.model = model

    def get_price(self):
        return self.price

    def set_price(self, price):
        self.price = price

    def get_car_type(self):
        return self.car_type

    def set_car_type(self, car_type):
        self.car_type = car_type

    def display_info(self):
        super().display()
        print("Name of the Car : ",self.name)
        print("Make Year of the Car : ",self.make_year)
        print("Color of the Car : ",self.color)
        print("Model of the Car :", self.model)
        print("Price of the Car :", self.price)
        print("Car Type of the Car :", self.car_type)

car1 = Car("Honda", 2022, "Red", "Civic", 200000, "Sedan")
car1.display_info()
car1.get_color()
car1.set_color("Black")
car1.display_info()
```

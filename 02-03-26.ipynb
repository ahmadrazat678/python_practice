```python
class area:
    def __init__ (self, a, b):
        self.a = a
        self.b = b

    def area(self):
        return self.a * self.b
    
class triangle(area):
    def __init__(self, a, b,):
        super().__init__(a, b)

    def area(self):
        c = super().area()
        ta = c/2
        return ta

t = triangle(15, 58)
print(t.area())
```

```python
 #Customer discount example using polymorphism

class Customer:
    def __init__(self, name):
        self.name = name

    def get_discount(self, amount):
        """Return final payable amount after discount."""
        raise NotImplementedError("Subclasses must implement get_discount")

class RegularCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.95  # 5% discount

class PremiumCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.90  # 10% discount

class VIPCustomer(Customer):
    def get_discount(self, amount):
        return amount * 0.80  # 20% discount

# demonstration
customers = [
    RegularCustomer("Alice"),
    PremiumCustomer("Bob"),
    VIPCustomer("Charlie")
]

for customer in customers:
    print(f"{customer.name}: ${customer.get_discount(100):.2f}")

amount = 100
for cust in customers:
    final = cust.get_discount(amount)
    print(f"{cust.__class__.__name__} {cust.name} pays: {final}")
```

```python
# create a payment class with a method called payamount
class Payment:
    def pay(self, amount):
        """Process a payment of the specified amount."""
        self.amount = amount
        return self.amount

# Subclasses 

class CreditCardPayment(Payment):
    def pay(self, amount):
        super().pay(amount)
        print("Credit card payment processed:", amount)

class JazzCashPayment(Payment):
    def pay(self, amount):
        super().pay(amount)
        print("JazzCash payment completed:", amount)

class BankTransferPayment(Payment):
    def pay(self, amount):
        super().pay(amount)
        print("Bank transfer successful:", amount)

p = CreditCardPayment()
p.pay(50000)

j = JazzCashPayment()
j.pay(1000)

b = BankTransferPayment()
b.pay(100000)
```

```python

# Employee salary calculation using polymorphism
class Employee:
    def calculate_salary(self):
        raise NotImplementedError("Subclasses must implement calculate_salary")

class FullTimeEmployee(Employee):
    def __init__(self, salary):
        self.salary = salary

    def calculate_salary(self):
        return self.salary

class PartTimeEmployee(Employee):
    def __init__(self, hourly_rate, hours):
        self.hourly_rate = hourly_rate
        self.hours = hours

    def calculate_salary(self):
        return self.hourly_rate * self.hours

class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend

    def calculate_salary(self):
        return self.stipend

# demonstrate polymorphism
employees = [
    FullTimeEmployee(5000),
    PartTimeEmployee(20, 120),
    Intern(1000)
]

for emp in employees:
    print(f"{emp.__class__.__name__} salary: {emp.calculate_salary()}")
```

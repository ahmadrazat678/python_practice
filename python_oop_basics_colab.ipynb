# Python OOP Basics — Class, Object, Constructor, self, Attributes, Methods

## 1) Class
**Explanation (2 lines):**  
A class is a blueprint for creating objects. It defines the properties (data) and behaviors (methods) that objects will have.

```python
# 10 Real‑World Class Examples
class Car: pass
class Student: pass
class Mobile: pass
class Book: pass
class BankAccount: pass
class Employee: pass
class Laptop: pass
class Dog: pass
class Hospital: pass
class Airplane: pass
```

## 2) Object
**Explanation (2 lines):**  
An object is an instance of a class. It represents a real‑world entity created from the class blueprint.

```python
car1 = Car()
student1 = Student()
phone1 = Mobile()
book1 = Book()
account1 = BankAccount()
emp1 = Employee()
laptop1 = Laptop()
dog1 = Dog()
hospital1 = Hospital()
plane1 = Airplane()
```

## 3) Constructor (__init__)
**Explanation (2 lines):**  
A constructor is a special method that runs automatically when an object is created. It initializes the object's data.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

class Student:
    def __init__(self, name):
        self.name = name

class Mobile:
    def __init__(self, model):
        self.model = model
```

## 4) self
**Explanation (2 lines):**  
self refers to the current object of the class. It is used to access the object's attributes and methods inside the class.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

car1 = Car("Toyota")
print(car1.brand)
```

## 5) Data / Attribute
**Explanation (2 lines):**  
Attributes are variables that belong to an object. They store the object's data or state.

```python
class Student:
    def __init__(self):
        self.name = "Ali"
        self.age = 20

s1 = Student()
print(s1.name, s1.age)
```

## 6) Function / Method
**Explanation (2 lines):**  
A method is a function defined inside a class. It describes the behavior or actions an object can perform.

```python
class Dog:
    def bark(self):
        print("Dog is barking")

d1 = Dog()
d1.bark()
```

```python
class BankAccount:
    
    def __init__(self):
        self.name = input("Enter your name: ")
        self.pin = None
        self.balance = 0
    # MENU INSIDE CLASS
    def menu(self):
        while True:
            print("\n===== BANK MENU =====")
            print("1. Set PIN")
            print("2. Deposit")
            print("3. Check Balance")
            print("4. Withdraw")
            print("5. Exit")
            
            choice = input("Enter choice: ")
            
            if choice == "1":
                self.set_pin()
            
            elif choice == "2":
                self.deposit()
            
            elif choice == "3":
                self.check_balance()
            
            elif choice == "4":
                self.withdraw()
            
            elif choice == "5":
                print(" Thank you for using our bank")
                break
            
            else:
                print(" Invalid choice")
    # Set PIN
    def set_pin(self):
        self.pin = input("Set your PIN: ")
        print(" PIN set successfully")
    
    # Deposit
    def deposit(self):
        amount = float(input("Enter amount to deposit: "))
        if amount > 0:
            self.balance += amount
            print(" Deposited:", amount)
        else:
            print(" Invalid amount")
    
    # Check Balance
    def check_balance(self):
        entered_pin = input("Enter PIN: ")
        if entered_pin == self.pin:
            print(" Your Balance is:", self.balance)
        else:
            print(" Incorrect PIN")
    
    # Withdraw
    def withdraw(self):
        entered_pin = input("Enter PIN: ")
        if entered_pin != self.pin:
            print(" Incorrect PIN")
            return
        
        amount = float(input("Enter amount to withdraw: "))
        
        if amount > self.balance:
            print(" Insufficient balance")
        elif amount <= 0:
            print(" Invalid amount")
        else:
            self.balance -= amount
            print(" Withdrawn:", amount)
            print(" Remaining Balance:", self.balance)
    
    
```

```python
account = BankAccount()
account.menu()
```

## 🔹 1. Import NumPy
NumPy (Numerical Python) is used for fast mathematical operations and array handling.

```python
import numpy as np
```

## 🔹 2. Creating NumPy Arrays
We can create arrays from lists, built-in methods, and random functions.


```python
# 5 Examples
np.array([1,2,3])
np.array([[1,2],[3,4]])
np.zeros((2,3))
np.ones((3,3))
np.arange(0,10,2)
```

### Random Arrays (5 Examples)

```python
np.random.rand(3)
np.random.rand(2,2)
np.random.randint(1,10,5)
np.random.randn(3)
```

## 🔹 3. Array Attributes & Methods
These help us understand array properties.


```python
arr = np.array([[1,2,3],[4,5,6]],dtype=float)
arr.shape
arr.ndim
arr.size
arr.dtype
arr.reshape(3,2)
```

```python
# max, min, argmax, argmin examples
arr.max()
arr.min()
arr.argmax()
arr.argmin()
arr.flatten()
```

## 🔹 4. Operations on Arrays
Copying, append, insert, sort, delete, combine, split.


```python
a = np.array([1,2,3])
b = np.array([4,5,6])
a.copy()
np.append(a,4)
np.insert(a,1,10)
np.sort(b)
np.delete(a,1)
```

```python
# Combining (5 Examples)
np.concatenate((a,b))
np.vstack((a,b))
np.hstack((a,b))
np.column_stack((a,b))
```

```python
# Splitting (5 Examples)
arr2 = np.arange(16).reshape(4,4)
np.split(np.array([1,2,3,4]),2)
np.array_split(np.array([1,2,3,4]),3)
np.vsplit(arr2,2)
np.hsplit(arr2,2)
```

## 🔹 5. Data Loading & Saving
NumPy allows saving and loading arrays.


```python
arr = np.array([[1,2,3],[4,5,6]])
np.save('array.npy', arr)
np.load('array.npy')
np.savetxt('array.txt', arr)
np.loadtxt('array.txt')
np.savez('multiple_arrays.npz', a=a, b=b)
```

## 🔹 6. NumPy Indexing & Selection
Accessing elements in 2D arrays and logical filtering.


```python
arr = np.array([[1,2,3],[4,5,6]])
arr[0]
arr[1,2]
arr[:,1]
arr[0:2,1:]
arr[-1]
```

```python
# Logical Selection (5 Examples)
arr[arr>3]
arr[arr%2==0]
arr[(arr>2)&(arr<6)]
arr[arr!=3]
arr[arr==5]
```

## 🔹 7. Broadcasting
Broadcasting allows operations between arrays of different shapes.


```python
a = np.array([1,2,3])
a + 5
a * 2
np.ones((3,3)) + np.array([1,2,3])
np.arange(3).reshape(3,1) + np.arange(3)
a - 1
```

## 🔹 8. Type Casting
Changing data types of arrays.


```python
a = np.array([1,2,3])
a.astype(float)
np.array([1.5,2.6]).astype(int)
np.array(['1','2']).astype(int)
np.array([True,False]).astype(int)
np.array([1,2,3], dtype='float32')
```

## 🔹 9. Arithmetic Operations
Basic mathematical operations on arrays.


```python
a = np.array([1,2,3])
b = np.array([4,5,6])
a + b
a - b
a * b
a / b
a ** 2
```

## 🔹 10. Universal Functions (Ufuncs)
Built-in mathematical functions.


```python
a = np.array([1,2,3])
np.sqrt(a)
np.exp(a)
np.sin(a)
np.log(a)
np.max(a)
```

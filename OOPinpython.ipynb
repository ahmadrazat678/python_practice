```python
class Animal():
    # using encapsulation:
    def __init__(self, name=None, species=None):
        if name is None:
            self.name = input("Enter the animal's name: ")
        else:
            self.name = name
        if species is None:
            self.__species = input("Enter the animal's species: ")
        else:
            self.__species = species

    def get_species(self):
        return self.__species

    def set_species(self, species):
        self.__species = species

    def sound(self):
        print("Animal makes a sound")

# inheritance: 
class Dog(Animal):
    def __init__(self, name=None, species=None, breed=None):
        super().__init__(name, species)
        if breed is None:
            self.breed = input("Enter the Dog's Breed: ")
        else:
            self.breed = breed

    # polymorphism: 
    def sound(self):
        print("Woof Woff")

# inheritance:
class Cat(Animal):
    def __init__(self, name=None, species=None, color=None):
        super().__init__(name, species)
        if color is None:
            self.color = input("Enter the Cat's Color: ")
        else:
            self.color = color

    # polymorphism: 
    def sound(self):
        print("Meow Meow")

animal_type = input("Enter the type of animal (Dog/Cat): ")

if animal_type.lower() == "dog":
    animal1 = Dog()

elif animal_type.lower() == "cat":
    animal2 = Cat()

else:
    print("Invalid animal type entered.")
    exit()


if animal_type.lower() == "dog":
    print("Animal Details: ")
    print("User entered: Dog")
    print("Name: ", animal1.name)
    print("Species: ", animal1.get_species())
    print("Breed: ", animal1.breed)
    print("animal1 make a sound: ", animal1.sound(""))

elif animal_type.lower() == "cat":
    print("Animal Details: ")
    print("User entered: Cat")
    print("Name: ", animal2.name)
    print("Species: ", animal2.get_species())
    print("Color: ", animal2.color)
    print("animal2 make a sound: ", animal2.sound())
    # example polymorphic call (uncomment to use):
    # animal2.sound()
```
